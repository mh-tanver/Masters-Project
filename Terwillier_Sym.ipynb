```python
G = SymmetricGroup(5)
G
```

```python
CL = [ccl.list() for ccl in G.conjugacy_classes()]
```

```python
gs = []
for ccl in G.conjugacy_classes():
    gs += ccl
```

```python
LA = []
for i in range(len(CL)):
    A = matrix(len(gs), len(gs))
    for j in range(len(gs)):
        for k in range(len(gs)):
            if gs[j]*gs[k].inverse() in CL[i]:
                A[j,k] = 1
    LA.append(A)
```

```python
def zero_position(A):
    for i in range(A.nrows()):
        for j in range(A.ncols()):
            if A[i,j] == 1:
                return i, j
    return None  
index = [zero_position(A) for A in LA] 
```

```python
def find_pijk(LA, index):
    d = len(LA)
    T = []
    for i in range(d):
        Pij = []
        for j in range(d):
            Aij = LA[i]*LA[j]
            pij = [Aij[idx[0], idx[1]] for idx in index]
            Pij.append(pij)
            
        mat_pij = matrix(Pij)
        T.append(mat_pij)
    return T


        
def dimension_t(T):
    s = T[0].nrows()
    d = len(T)
    b = []
    for i in range(s):
        bi = matrix(d,d)
        for j in range(d):
            for k in range(d):
                bi[j,k] = T[k][i,j]
        b.append(bi)
    return b
```

```python
T = find_pijk(LA, index)
```

```python
b = dimension_t(T)
b
```

```python
def pdiagonal(B):
    p = B[0].eigenmatrix_right()[1]
    for i in range(1, len(B)):
        Dtemp = p.inverse()*B[i]*p
        pt = Dtemp.eigenmatrix_right()[1]
        p = p*pt
    return p

def find_M(P,B):
    pi = p.inverse()
    M = matrix(pi.base_ring(), p.nrows(), p.ncols())
    
    for i in range(len(B)):
        Bi= B[i]
        V = pi*Bi*p
        M[:, i] = (matrix(V.diagonal())).transpose()[:, 0]     
    return M

def find_idem(M, LA):
    Mi = M.inverse()
    return [sum([Mi[i,j]*LA[i] for i in range(len(LA))]) for j in range(len(LA))]
```

```python
def basis_terwil(LA, LE):
    d = len(LA)
    S = []
    for i in range(d):
        for j in range(d):
            for k in range(d):
                for l in range(d):
                    for m in range(d):
                        ea = LE[i]* LA[j] * LE[k]* LA[l]*LE[m]
                        if ea != 0:
                            S.append(ea)
    return S
```
